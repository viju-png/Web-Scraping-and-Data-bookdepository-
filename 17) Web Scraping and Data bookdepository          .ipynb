# Web Scraping and Data Preparartion                     

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url="https://www.bookdepository.com/"
response=requests.get(url)
```

```python
html=response.content
html
```

```python
soup=bs(html,'lxml')
soup
```

```python
# get the book title
all_h3=soup.find_all("h3",class_="title")
for h3 in all_h3:
    title=h3.get_text(strip=True)
    print(title)
```

```python
# geting book format
import pandas as pd
formats=soup.select("div.item-info p.format")
formats_series=pd.Series(formats)
formats_series.value_counts()
```

```python
#getting publication date
date=soup.find_all("p",class_="published")
date=[date.get_text()[-4:]for date in date]
date_series=pd.Series(date)
date_series.value_counts()
```

```python
# getting the find the price
final_prices=[]
prices=soup.find_all("p",class_="price")
for price in prices:
    origanal_price=price.find("span",class_="rrp")
    if origanal_price:
        current_price=str(origanal_price.previousSibling).strip()
        current_price=current_price.split("€")[0].replace(",",".")
    else:
        current_price=price.get_text(strip=True).split("€")[0].replace(" ,",".")
    final_prices.append(current_price)
```

```python
book_final_price=pd.DataFrame(final_prices,columns=["price"])
```
